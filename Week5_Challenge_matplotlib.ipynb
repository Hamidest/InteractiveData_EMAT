```python
# Step 1: Import
import matplotlib.pyplot as plt
import pandas as pd
```

# Step 2: Choose Style
Pie chart and bar chart have been chosen in order to work with in this dataset.
for the Pie chart I would like to show the "sex" variables of participants
for the bar chart I would like to show the "Evaluation" parameter and to what extend they were agree to the given term.

```python
# Step 3: Importing CSV
df=pd.read_csv("cces_example_data.csv")
df=pd.DataFrame(df)
df.head()
```

```python
# Step 3: cleaning the data and categorize variables
df_clean=df.astype({"sex":'category',"pid3":'category',"nonwhite":'category',"Evaluation":'category'})
```

```python
# Step 4: Create and show in your notebook an empty figure that contains two axes 
fig1 , ax1 = plt.subplots(1,2)
ax1[0].set_title("Age Histogram")
ax1[1].set_title("Political Identity")
```

```python
#Step 5:
# bar chart
fig2, ax2=plt.subplots()
# clean
df_ev=df_clean.groupby('Evaluation').count().caseid
ax2.bar([0,1,2,3], df_ev)
# try set_xticks
ax2.set_xticks([0,1,2,3],["Strongly Disagree", "Disagree","Agree","Strongly Agree"])
```

```python
# Step 6:
fig3 ,ax3=plt.subplots()
ax3.scatter(df_clean['ConMediaUse'],df_clean['ImmiPhobia'],alpha=.07)
plt.show()
```

```python
# additional steps
fig4, ax4,=plt.subplots()
df_sex=df_clean.groupby("sex")
ax4.pie(df_sex.count().caseid,labels=["male","female"],autopct='%1.1f%%')
ax4.set_title("Sex")
```
